### 1. Write a Python function to sum all the numbers in a list.
####  check the function with [8, 2, 3, 1, 7,5,6]

```python
my_list= [8, 2, 3, 1, 7, 5, 6]

def sum(numbers):
    total=0 # initalized total
    for i in numbers:
        total=total+i
    return total

sum(my_list)
```

### 2. Write a Python function to multiply all the numbers in a tuple.
#### Check the function with (8, 2, 3, 1, 7,5,6)

```python
my_tuple=(8, 2, 3, 1, 7,5,6)

def multiply(numbers):
    total=1 # initalized total
    for i in numbers:
        total*=i
    return total

multiply(my_tuple)
```

### 3. Write a Python program to print the even numbers from a given 10 number list.
[1, 2, 3, 4, 5, 6, 7, 8, 9]

```python
list2=[1, 2, 3, 4, 5, 6, 7, 8, 9]
def even(List):
    even_numbers=[]
    for i in List:
        if i%2==0:
            even_numbers.append(i)
    return even_numbers
            
```

```python
even(list2)
```

```python
list2=[1, 2, 3, 4, 5, 6, 7, 8, 9]
def even_or_odd(List):
    even_numbers=[]
    odd_numbers=[]
    for i in List:
        if i%2==0:
            even_numbers.append(i)
        else:
            odd_numbers.append(i)
    print(f"Even_numbers: {even_numbers}")
    print(f"Odd_numbers: {odd_numbers}")
```

```python
even_or_odd(list2)
```

```python
list2=[1, 2, 3, 4, 5, 6, 7, 8, 9]
def multiple_of_3(List):
    numbers=[]
    for i in List:
        if i%3==0:
            numbers.append(i)
    return numbers
```

```python
multiple_of_3(list2)
```
