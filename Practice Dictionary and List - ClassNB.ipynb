### Convert the lists to dictionary

```python
keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]
```

### Find who is younger
Dict9 = {
  'Sam': 82,
  'Harry': 65,
  'Jack': 75
}

```python
Dict9 = {'Sam': 82, 'Harry': 65, 'Jack': 75}

younger=min(Dict9.values())
print(younger)

for k in Dict9:
   # print(Dict9[k])
    if Dict9[k]==younger:
        print(f"The youngest person is {k}. He is {younger} years old.")
```

```python
oldest=max(Dict9.values())

for x in Dict9:
    if Dict9[x]==oldest:
        print(f"The oldest person is {x}. He is {oldest} years old.")
```

### Write a Python script to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x**3). 
- Sample Dictionary ( n = 5) :
- Expected Output : {1: 1, 2: 8, 3: 27, 4: 64, 5: 125}

```python
n=int(input("Enter the value of n: "))
Dict={}
for x in range(1,n+1):
    #print(x)
    #print({x:x**3})
    Dict.update({x:x**3})
print(Dict)
```
