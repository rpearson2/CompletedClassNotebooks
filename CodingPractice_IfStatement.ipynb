### A student will not be allowed to sit in exam if his/her attendence is less than 75%.
- Take following input from user
    - Number of classes held
    - Number of classes attended.
    - And print percentage of class attended
- Is student is allowed to sit in exam or not.

```python
# Homework
classes_held=int(input("Enter the total number of classes held: "))
classes_attended=int(input("Enter the total number of classes attended: "))
percent=classes_attended/classes_held
print(f"The student's attendence is {percent*100}%.")
if percent<.75 and percent>=0:
    print(f"The sudent's attendence is {percent*100}%. They are not allowed to sit in the exam because it is less than 75%.")
elif percent<0 or percent>1:
    print("There is an error in your input, attendence can't be less than 0% or greater than 100%.")
else:
    print(f"The sudent's attendence is {percent*100}%. They are allowed to sit in the exam because it is greater than 75%.")
```

### Write a program that accepts the lengths of three sides of a triangle as inputs. The program output should indicate whether or not the triangle is a right triangle (Recall from the Pythagorean Theorem that in a right triangle, the square of one side equals the sum of the squares of the other two sides

```python
side1=int(input("Enter the first side length of the triangle: "))
side2=int(input("Enter the second side length of the triangle: "))
side3=int(input("Enter the third side length of the triangle: "))
if side1**2==side2**2 + side3**2:
    print("It is a right triangle.")
elif side2**2==side1**2 + side3**2:
    print("It is a right triangle.")
elif side3**2==side1**2 + side2**2:
    print("It is a right triangle.")
else:
    print("It is not a right triangle.")
```

## The given number is of one digited or two digited or three digited or more than three digited.

```python
number=int(input("Enter a number: "))
if number >=0 and number <=9 or number >=-9 and number <=0:
    print(f"The number {number} is one digit.")
elif number >9 and number <=99 or number>=-99 and number<-9:
    print(f"The number {number} is two digits.")
elif number >99 and number <=999 or number >=-999 and number <-99:
    print(f"The number {number} is three digits.")
else:
    print(f"The number {number} is more than three digits.")
```

### Check the input character is an uppercase letter or lowercase letter or a digit or a special character


```python
char=input("Enter an input character: ")
if ord(char)>=65 and ord(char)<=90:
    print(f"The character {char} is an uppercase letter.")
elif ord(char)>=97 and ord(char)<=122:
    print(f"The character {char} is a lowercase letter.")
elif ord(char)>=48 and ord(char)<=57:
    print(f"The character {char} is a digit.")
else:
    print(f"The character {char} is a special character.")
```

## The given year is a leap year or not
To determine whether a year is a leap year, follow these steps:

1. If the year is evenly divisible by 4, go to step 2. Otherwise, go to step 5.
2. If the year is evenly divisible by 100, go to step 3. Otherwise, go to step 4.
3. If the year is evenly divisible by 400, go to step 4. Otherwise, go to step 5.
4. The year is a leap year (it has 366 days).
5. The year is not a leap year (it has 365 days).
* Find out for the year 1700, 1800, 2000, 2400

```python
# fix
year=int(input("Enter a year: "))

if (year%4==0 and year%100!=0) or year%400==0:
    print(f"The year {year} is a leap year.")
else: 
    print(f"The year {year} is not a leap year")
```


Write a program that reads a, b, c, d, k and x and prints the value of f(x)

```python
a=int(input("Enter the value of a: "))
b=int(input("Enter the value of b: "))
c=int(input("Enter the value of c: "))
d=int(input("Enter the value of d: "))
k=int(input("Enter the value of k: "))
x=int(input("Enter the value of x: "))

if x>k:
    fx=a*(x**3)-b*(x**2)+c*(x)-d
elif x==k:
    fx=0
elif x<k:
    fx=-a*(x**3)+b*(x**2)-c*(x)+d
print(fx)
```
