## Math modules
Python has a built-in module that you can use for mathematical tasks.
The math module has a set of methods and constants.
- math.floor(x) largest integer not greater than x
- math.ceil(x) Smallest integer not less than x
- math.sqrt(x) square root of any number
- round() float point number from the decimal value to the closest multiple of 10
- math.acos()	Returns the arc cosine of a number
- math.gcd()	Returns the greatest common divisor of two integers
- math.pow()	Returns the value of x to the power of y
- math.pi	Returns PI (3.1415...)
- math.tau	Returns tau (6.2831...)
- More math modules- https://www.w3schools.com/python/module_math.asp

```python
import math
print("pi:",math.pi)
```

```python
PI=math.pi
print("floor: ",math.floor(PI))
print("Ceiling: ",math.ceil(PI))
print("Round: ", round(PI))
print("Pi to the Power of 2: ", math.pow(PI,2))
print("Pi to the power of 5: ",round(math.pow(PI,5)))
print("Square root of pi: ", math.sqrt(PI))
```

## Heron's formula for area of triangle


```python
import math
a=int(input("Enter a: "))
b=int(input("Enter b: "))
c=int(input("Enter c: "))

s= (a+b+c)/2

area=math.sqrt(s*(s-a)*(s-b)*(s-c))
print("Area of the triange is: ", area)
```

```python
import math
a=int(input("Enter a: "))
b=int(input("Enter b: "))
c=int(input("Enter c: "))
root=math.sqrt(math.pow(b,2)-4*a*c)
x1=(-b+root)/(2*a)
x2=(-b-root)/(2*a)
print(f"Root 1: {x1} and Root 2: {x2}")
print(f"Rounded Root 1: {round(x1)} and Rounded Root 2: {round(x2)}") # rounds to whole number
print(f"Rounded Root 1 to 3 dec. places: {x1:.3f} and Rounded Root 2 to 3 dec. places: {x2:.3f}") # rounds to 3 decimal places
```
