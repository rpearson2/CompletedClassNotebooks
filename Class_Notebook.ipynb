### Create the person class

```python
class Person():
    # init method, required to create a class
    def __init__(self, Name, Sex, Profession, Study_hr):
        self.Name= Name
        self.Sex= Sex
        self.Profession=Profession
        self.Study_hr=Study_hr
        
    # work function/method
    def work(self):
        print(f"{self.Name} is a/an {self.Profession}.")
    
    # study function/method
    def study(self):
        print(f"{self.Name} studied for {self.Study_hr} hours.")
```

```python
# object 1
Jessica=Person("Jessica","Female","Engineer",20)
# instance
Jessica.work()
```

```python
Jessica.study()
```

```python
# object 2
John=Person("John","Male","Doctor",15)
John.work()
John.study()
```

### Incrementing an Attribute’s Value Through a Method called incremental_odometer

```python
class Car():
    def __init__(self, make, model, year):
        self.make=make
        self.model=model
        self.year=year
        self.odometer_reading=0
        
    def get_descriptive_name(self):
        long_name= str(self.year) + " " + self.make + " " + self.model
        return long_name
    
    def read_odometer(self):
        print(f"The car has {self.odometer_reading} miles on it.")
        
    def update_odometer(self, miles):
        self.odometer_reading=miles
        
    def incremental_odometer(self,miles):
        self.odometer_reading+=miles
```

```python
my_new_car=Car("Honda","Civic",2020)
print(my_new_car.get_descriptive_name())
my_new_car.read_odometer()
my_new_car.update_odometer(100)
my_new_car.read_odometer()
my_new_car.incremental_odometer(200)
my_new_car.read_odometer()
```

```python
my_new_car.incremental_odometer(500)
my_new_car.read_odometer()
```

# Inheritance: Extending Classes To Make New Classes
- As you might know, creating a fully functional class in an object-oriented programming language is time-consuming because real classes perform a lot of complex tasks.

- In Python, you can get the features you want from an `existing class(parent)` to create a `new class(child)`. This Python feature is called inheritance. By inheritance, you can: 

1.  obtain the features of a parent class,
2.  change the features that you don’t need,
3.  add new features to your child class. (derived class or subclass)

### Instead of writing the same class again and again, we can define a parent class `Data_Professional` and 3 child classes of the Data_Professional class: Data_Analyst, Data_Scientist, and Data_Engineer.

## Parent class

```python
class Data_Professional():
    def __init__(self, name, SQL_experience):
        self.name=name
        self.SQL_experience=SQL_experience
        
    def know_math_stat(self):
        return True
    
    def know_programming(self):
        return True
```

## Child Classes

```python
class Data_Scientist_child(Data_Professional):
    def __init__(self, name, SQL_experience):
        super().__init__(name,SQL_experience) # gets name and SQL from its super class (parent class)
```

```python
# Object 1 for Data_Scientist
DS_1=Data_Scientist_child("John Cason",8)
print(f"{DS_1.name} has {DS_1.SQL_experience} years of SQL experience.")
print(f"Does he know math and stat? {DS_1.know_math_stat()}")
print(f"Does he know programming? {DS_1.know_programming()}")
```

```python
# chid class for data engineer
class Data_Engineer_child(Data_Professional):
    def __init__(self, name, SQL_experience):
        super().__init__(name, SQL_experience) # get the name and sql experience method from parent class
```

```python
# object 2 for data engineer
DE_1=Data_Engineer_child("John Doe",10)
print(f"{DE_1.name} has {DE_1.SQL_experience} years of SQL experience.")
print(f"Does he know math and stat? {DE_1.know_math_stat()}")
print(f"Does he know programming? {DE_1.know_programming()}")
```

### An empty child class has the properties of its parent class

```python
class Data_Analyst_child(Data_Professional):
    pass
# empty class also inherites all the properties/methods from the parent class
DA_1=Data_Analyst_child("John Willy 2",7)
print(f"{DA_1.name} has {DA_1.SQL_experience} years of SQL experience.")
print(f"Does he know math and stat? {DA_1.know_math_stat()}")
print(f"Does he know programming? {DA_1.know_programming()}")
```

## Overriding Classes

```python
# child class called computer scientist
class computer_scientist(Data_Professional):
    def __init__(self,name, SQL_experience):
        super().__init__(name,SQL_experience)
    # overwriting the parent class
    def know_programming(self):
        return False
    
#object
CS_1=computer_scientist("John Johnson", 10)

print(f"{CS_1.name} has {CS_1.SQL_experience} years of SQL experience.")
print(f"Does he know math and stat? {CS_1.know_math_stat()}")
print(f"Does he know programming? {CS_1.know_programming()}")
```

### Practice: Create a class named `Polygon`. with a `side_length` properties, and a perimerter method:


```python
class Polygon():
    def __init__(self, side_length):
        self.side_length=side_length
    
    def perimeter(self):
        return sum(self.side_length)
```

```python
some_shape=Polygon([1,2,3,4,5])
print("Perimeter: ",some_shape.perimeter())
```

```python
some_shape_1=Polygon([7,8,9,12,67,98,6])
print("Perimeter: ",some_shape_1.perimeter())
```

### Create a child called `Triangle` where the child inherit the `perimeter` and has its own area method

```python
import math
class Triangle(Polygon):
    def __init__(self, side_length):
        super().__init__(side_length)
    # perimeter method will be inheriting from parent
    # create area method for triangle using Heron's formula
    def area(self):
        a,b,c=self.side_length
        s=(a+b+c)/2
        return math.sqrt((s*(s-a)*(s-b)*(s-c)))
```

```python
a_triangle=Triangle([3,4,5])
print("Perimeter:",a_triangle.perimeter())
print("Area:", a_triangle.area())
```

```python
a_rectangle=Rectangle([2,3,2,3])
print("Perimeter:",a_rectangle.perimeter())
print("Area:", a_rectangle.area())
```

## Create a simple calculator using classes

```python
class Calculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    
    def addition(num1,num2):
        print("Result: ", num1+num2)
        
    def subtraction(num1,num2):
        print("Result: ",num1-num2)
    
    def multiplication(num1, num2):
        print("Result: ", num1*num2)
        
    def division(num1,num2):
        print("Result: ", num1/num2)
        
while True:
    print("Select a number: ")
    print("Enter '1' for addition")
    print("Enter '2' for subtraction")
    print("Enter '3' for multiplication")
    print("Enter '4' for division")
    print("Enter '5' to quit")
    choice=int(input("Enter your choice (1-5): "))
    
    if choice==1:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Calculator.addition(num1,num2)
    elif choice==2:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Calculator.subtraction(num1,num2)
    elif choice==3:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Calculator.multiplication(num1,num2)
        
    elif choice==4:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        if num2==0:
            print("Infinity")
        else:
            Calculator.division(num1,num2) 
    
    elif choice==5:
        break
        
    else:
        print("Invalid number, please choose a number between (1-5)")
```

## ATM machine 

```python
class ATM():
    def __init__(self):
        self.balance=0
    
    def deposit(self):
        amount=int(input("Enter the amount you want to deposit: "))
        self.balance+=amount
        print(f"The amount deposited: {amount}")
        
    def balance_check(self):
        print(f"The net avalialable balance is {self.balance}.")
        
    def withdrawn(self):
        amount=int(input("Enter the amount you want to withdraw: "))
        
        if self.balance>amount:
            self.balance-=amount
            print(f"You withdraw {amount}.")
        else:
            print("Insufficient balance")
```

```python
my_account=ATM()
my_account.balance_check()
my_account.deposit()
```
